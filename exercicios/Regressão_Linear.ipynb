```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Regress√£o Linear

## Carregando dataset

```python
from sklearn.datasets import fetch_openml

housing = fetch_openml(name="house_prices", as_frame=True)
X = housing.frame.drop(columns=["SalePrice"])
y = housing.frame["SalePrice"]

plt.scatter(X["GrLivArea"], y)
plt.xlabel("Above Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.title("Scatter Plot of Above Ground Living Area vs. Sale Price")
plt.show()
```

## Implementando a regress√£o linear

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X[["GrLivArea"]], y)

y_pred = regressor.predict(X[["GrLivArea"]])

plt.scatter(X["GrLivArea"], y, label="Data Points")
plt.xlabel("Above Ground Living Area (sq ft)")
plt.ylabel("Sale Price")
plt.title("Scatter Plot with Linear Regression Line")

plt.plot(X[["GrLivArea"]], y_pred, color='red', linewidth=2, label="Linear Regression Line")

plt.legend()
plt.show()
```
